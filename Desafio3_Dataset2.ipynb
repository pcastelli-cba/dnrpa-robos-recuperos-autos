#Leyendo archivos para generar dataset

Importar archivos de GIT

```python
import pandas as pd
#anios = ['2018','2019','2020','2021','2022']
anios = ['2020','2021','2022']
meses = ['01','02','03','04','05','06','07','08','09','10','11','12']

def listado(anios,meses):
  lista=[]
  for a in range(len(anios)):
    for m in range(len(meses)):
      anio=anios[a]
      mes=meses[m]
      print("Comprobando datos de Año : "+str(anios[a])+" Mes : "+str(meses[m]))
      try :
        url = 'https://raw.githubusercontent.com/pcastelli-cba/dnrpa-robos-recuperos-autos/main/dnrpa-robos-recuperos-autos-'+str(anio)+str(mes)+'.csv'
        df = pd.read_csv(url, index_col=0)
        lista.append(url)
      except : 
        print("Datos inexistentes para el Año : "+str(anios[a])+" Mes : "+str(meses[m]))
        break
  return lista

dataFrame = pd.concat(map(pd.read_csv, listado(anios,meses)), ignore_index=True)
print(dataFrame)

dataFrame.to_csv('archivo unificado.csv')

```

#DESAFIO 3 - ESTRUCTURANDO UN PROYECTO DS PARTE 1

Generamos un nuevo dataset acumulando las cantidades por mes de autos robados por marca

```python
df_marca=dffiltro.groupby(['mes','automotor_marca_descripcion']).size().iloc[2:].reset_index().pivot(columns='automotor_marca_descripcion', index='mes', values=0)
df_marca.fillna(0, inplace=True)
df_marca
```

```python
#Seleccionamos las marcas con mas robos de acuerdo a los graficos vistos en el desafio anterior
df_top_marca=df_marca[['TOYOTA','VOLKSWAGEN','FIAT','RENAULT','CHEVROLET','FORD','PEUGEOT','CITROEN']]
df_pct_top_marca=df_top_marca.pct_change()
df_top_marca
```

```python
df_top_marca.plot(kind='bar', figsize=(20,12), ylabel='Frecuencia', title='Meses')
```

Podemos observar en el grafico que los ultimos 3 meses del año se registran menos cantidad de robos de autos, esta tendencia es seguida por todas las marcas seleccionadas
